```python
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import classification_report

# Load the dataset
dataset = dataset_imputed

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(dataset.drop('outcome', axis=1), 
                                                    dataset['outcome'], test_size=0.3, 
                                                    stratify=dataset['outcome'], 
                                                    random_state=42)

# Apply oversampling to the training data
oversampler = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = oversampler.fit_resample(X_train, y_train)

# Select top 10 features using a decision tree classifier
dt = DecisionTreeClassifier(random_state=42)
selector = SelectFromModel(dt, max_features=10)
selector.fit(X_train_resampled, y_train_resampled)

# Apply feature selection to the training and testing data
X_train_selected = selector.transform(X_train_resampled)
X_test_selected = selector.transform(X_test)

# Train a decision tree model on the resampled and selected data
dt = DecisionTreeClassifier(class_weight='balanced', random_state=42)
dt.fit(X_train_selected, y_train_resampled)

# Make predictions on the testing data and adjust the decision threshold
y_pred_prob = dt.predict_proba(X_test_selected)[:, 1]
y_pred_threshold = (y_pred_prob > 0.4).astype(int)
accuracy = accuracy_score(y_test, y_pred_threshold)
auc = roc_auc_score(y_test, y_pred_threshold)
# Print the classification report
print("AUC score:", auc)
print(f"Accuracy: {accuracy:.4f}")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_threshold))
print("Classification Report:\n", classification_report(y_test, y_pred_threshold))

```
